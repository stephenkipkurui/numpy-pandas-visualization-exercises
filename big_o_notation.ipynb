```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = range(-3, 4)

y1 = [n + 1  for n in x]

y2 = [np.log(n) for n in x]

y3 = [n for n in x]

y4 = [(np.log(n) * n) for n in x]

y5 = [np.prod(n*n) for n in x]

y6 = [2**n for n in x]

#y7 = [n! for n in x]

y8 = [n**n for n in x]



plt.figure(figsize = (20, 9))

y = [n for n in x ]


plt.plot(x,y1, c='r', label='$\O(1)$')
plt.plot(x, y2, label='$\O(log n)$')
plt.plot(x, y3, c='g', label='$\O(n)$')
plt.plot(x, y4, label='$\O(n log n)$')
plt.plot(x, y5, label='$\O(n^2)$')
plt.plot(x, y6, label='$\O(2^n)$')
#plt.plot(x, y7, label='$\O(n!)$')
plt.plot(x, y8, label='$\O(n^n)$')

plt.title('Big O Notation', c='orange')
plt.xlabel('$Elements$', c='g')
plt.ylabel('$Operations$', c='g')

plt.legend(loc='upper right')

```
