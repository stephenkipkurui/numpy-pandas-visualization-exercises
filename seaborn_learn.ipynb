```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.dtypes
```

```python
tips
```

```python
sns.relplot(data = tips, x='total_bill', y='tip', kind='scatter', hue='day', col='day')
```

```python
sns.displot(data = tips, x='total_bill', col='sex', kind='hist', hue='smoker', rug=True)
```

```python
sns.dogplot()
```

```python
sns.pairplot(tips, hue='smoker')
```

```python
data = pd.crosstab(tips.time, tips['size'])
sns.heatmap(data, annot=True, cmap=plt.cm.Greens)
```

```python
tips.corr()
```

```python
x = list(range(150))

y = [n + 1 for n in x]

plt.plot(x, y)
```

```python
x = list(range(-20, 20))

y = [n**2 for n in x]

plt.annotate('Center', xy=(0,0), xytext=(0,50), arrowprops={'facecolor': 'blue'})

plt.plot(x, y, ls ='--')


```
