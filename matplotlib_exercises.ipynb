2). Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

You can use functions from the math module to help implement some of the equations above.

```python
#      y = √x 


x = list(range(-50, 50))

y = [(i**0.5) for i in x]


plt.plot(x, y)
```

```python
#     y = x^3

x = list(range(-50, 50))

y = [(c**3) for c in x]

plt.plot(x, y)

```

```python
#     y = 2x


x = list(range(-50, 50))

y = [(j**j) for j in x]


plt.plot(x, y)
```

```python
#     y = 1 / ( x + 1 )

x = list(range(-50, 50))

y = [1 / (t + 1) for t in x]

plt.plot(x, y)
```
